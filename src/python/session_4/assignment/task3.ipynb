1- Reversed word

```python
def reverse_words(s):
    words = s.split()
    return " ".join(words[::-1])
```

```python
reverse_words("The greatest victory is that which requires no battle")
```

2- Are You Playing Banjo?

```python
def are_you_playing_banjo(name):
    return True if name[0].lower() == "r" else False
```

```python
are_you_playing_banjo("rabbit")
```
