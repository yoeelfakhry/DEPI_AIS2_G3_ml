Two sum

```python
class Solution:
    def __init__(self , num:list , target:int):
        self.num    = num
        self.target =  target
    def twoSum(self):
        """
        Return the indices of two numbers in the list that sum to the target

        Parametes:
        num (list)  : The list you will search in
        target (int) : The target sum

        Returns:
        list : Two indices numbers
        """
        result = []
        for i in range(len(self.num)-1):
          for j in range (i+1 , len(self.num)):
            if self.num[i]+ self.num[j] == self.target :
                result.append(self.num.index(self.num[i]))
                result.append(self.num.index(self.num[j]))
        return result
        
new = Solution([2,3,4], 6)
new.twoSum()
```
