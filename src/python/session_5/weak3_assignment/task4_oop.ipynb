```python
class Factorail:
  """
  Task1 class the calculate the factorial of the number
  Attributes
  number (int) : Number that will calculated
  """
  def __init__(self , number:int):
    self.number = number

  def fact(self):
    """
    Calculating the factorial of the number
    (factorial is the multiplication from 1 to number)
    Return:
    int : The factorial number
    """
    if self.number == 1:
      return 1
    if self.number == 2:
      return 2
    return self.number * Factorail(self.number-1).fact()
```

```python
check = Factorail(5)
print(check.fact())
```

```python
class PrimeNumber:
  """
  Task2 Determine the number if it prime or not
  Attributes
  number (int) : Number that will checked
  """
  def __init__(self , number:int):
    self.number = number

  def prime(self):
    """
    Determine if number is prime or not
    (Prime number greater than 1 and have exactly two divisors 1 and itself)
    Return:
    bool : True if the number is prime and False if the number is not prime
    """
    if self.number <= 1:
      return False
    for i in range(2,self.number):
      if self.number % i ==0 :
        return False
    
    return True
```

```python

check = PrimeNumber(7)
print(check.prime())
```

```python
class Dividors:
    """
    Class to find common dividors of two numbers

    Attribures:
    num1 (int) : First number
    num2 (int) : second number
    """
    def __init__(self , num1 , num2):
        self.num1 = num1
        self.num2 = num2

    def dividors(self):
        """
        Find all common dividors if the two number
        
        Return:
        list : list of numbers
        """
        output = []
        for i in range(1 , max(self.num1 , self.num2)):
            if self.num1 % i ==0 and self.num2 % i ==0 :
                output.append(i)
        return output
    
```

```python
result = Dividors(10,20)
result.dividors()
```

```python
class LongestCommonSubsequence:
    """
     Finds the longest common subsequence between two strings.
    
    Attributes
     str1 (str) : first string
     str2 (str) : second string
     """
    def __init__(self , str1 , str2):
        self.str1 = str1
        self.str2 = str2

    def common(self):
        """
         Find longest common subsequence using a simple approach.
         Return 
         str : the sommon subdequence
         """
        str1_words = self.str1.split()
        str2_words = self.str2.split()
        output = []
        for i in str1_words :
            for j in str2_words:
                if i == j :
                    output.append(i)
        return " ".join (output)
```

```python
result = LongestCommonSubsequence("Welcome to Machine Learning Diploma" , "I am studying a Machine Learning Course" )
result.common()
```

```python
class ReverseWord:
    """
    class that reverse every word in the string

    Attribute:
    sentence (str) : Given sentence that i need to reverse 
    """
    def __init__(self , sentence):
        self.sentence = sentence
    
    def reverse_words(self):
        """
        Reverse words in the given sentence

        Return:
        str : Reversed words
        """
        words = self.sentence.split()
        return " ".join(words[::-1])
```

```python
result = ReverseWord("hello world")
result.reverse_words()
    
```

```python
import random
class PassGenerator:
    """
    class that Generate random password from characters with the given lenght
    Attributes:
    lenght (int) : the lenght of the password
    """
    def __init__(self , lenght:int):
        self.lenght = lenght

    def pass_generator(self):
        """
        Generate random password from characters with the given lenght
        Return:
        str : random password
        """
        characters = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#$%^&*"
        return "".join( characters[random.randrange(0,len(characters))] for i in range(self.lenght))
        # OR
        # return "".join( random.choice(characters) for i in range(lenght))
```

```python
result = PassGenerator(8)
result.pass_generator()
```

```python
class UniqueWords:
    """
    Class that count word occurence in the list
    Attributes
    words (list) : list of words that i need to count

    """
    def __init__(self , words):
        self.words = words

    def unique_words(self):
        """
        Counting word occurence in the list
        Return:
        dict : dictionary of key that contain words and values that contain it's count
        """
        uni_words = set(self.words)
        lst_words = list(uni_words)
        output= {}
        for word in range(len(lst_words)):
            output[lst_words[word]] = self.words.count(lst_words[word])
        return output
```

```python
result = UniqueWords(["Welcome", "Ali", "Hi", "Ali", "No", "Hi", "No", "Ali", "No", "Ali"])
result.unique_words()
```

```python
import random
class Chatbot:
    """
    class chatbot
    """
    def __init__(self , user_input):
        self.user_input = user_input

    def get_response(self):
        responses = {
        "hello": ["Hello!", "Hi there!", "Greetings!"],
        "how are you": ["I'm doing well, thank you!", "I'm fine, how about you?"],
        "goodbye": ["Goodbye!", "See you later!", "Farewell!"],
        "default": ["I'm sorry, I didn't understand.", "Could you please rephrase that?"]
        }
        for key in responses:
            if key in self.user_input:
                return random.choice(responses[key])
    
        return random.choice(responses["default"])

    def chatbot(self):
            print("Chatbot: Hi! How can I assist you today?")
            
            while True:
                self.user_input = input("User: ").lower()
                response = Chatbot(self.user_input).get_response()
                print("Chatbot:", response)
                
                if self.user_input == "goodbye":
                    break

```

```python
result = Chatbot("Hello!")
result.chatbot()
```
