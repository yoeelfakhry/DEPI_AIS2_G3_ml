```python
import os 
import random
#create path
path ="thanos_project"

if not os.path.exists(path):
    os.makedirs(path)
#create ten files
for i in range(1,11):
    file_path = os.path.join(path , "thanos_" + str(i) + ".txt")
    with open(file_path ,'w') as file :
        file.write("")

# count the files inside the folder
all_files = [f for f in os.listdir(path) if os.path.isfile(os.path.join(path, f))]
count_before = len(all_files)
print("Numbers of files before removing : " , count_before)
del_number = count_before // 2

if del_number > 0:
    # Get list of all files
    all_files = [f"thanos_{i}.txt" for i in range(1, 11)]
    # Randomly select files to delete
    files_to_delete = random.sample(all_files, del_number)
    
    for file_name in files_to_delete:
        file_path = os.path.join(path, file_name)
        os.remove(file_path)
        print(f"Deleted: {file_name}")
#count the remaining path
files_after = len([f for f in os.listdir(path) if os.path.isfile(os.path.join(path, f))])
print("Numbers of files after removing : " , files_after)


```
