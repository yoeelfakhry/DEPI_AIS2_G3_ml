```python
def sum(num_1, num_2):
    
    return num_1 + num_2

def sub(num_1, num_2):
    
    return num_1 - num_2
def div(num_1, num_2):
    
    return num_1 / num_2
def multply(num_1, num_2):
    
    return num_1 * num_2

def main():
    print("---------calc------")
    print("choose operations")
    print("1-sum")
    print("2-sub")
    print("3-div")
    print("4-mul")
    
    choice = input("Enter the operation")
    try :
        choice_num = int(choice)
        if choice_num < 1 or choice_num >4 :
            print("The operations number must be between 1 and 4")
            return
    except  ValueError:
        print("Enter numbers only")
        return

    try :
        num_1 = float (input("Enter first number"))
        num_2 = float (input("Enter second number"))
    except ValueError:
            print("Enter Valide number") 

    if choice_num == 3 and num_2 == 0:
        print("can not devide by zero")
        return
    
    if choice == "1":
        result = sum(num_1, num_2)

    elif choice == "2":
        result = sub(num_1, num_2)
        
    elif choice == "3":
        result = div(num_1, num_2)
        
    elif choice == "4":
        result = multply(num_1, num_2)
        
    else:
        print("invalid choice")
        return

    print("Result: ", result)
    
    
if __name__ == "__main__":
    main()
```
