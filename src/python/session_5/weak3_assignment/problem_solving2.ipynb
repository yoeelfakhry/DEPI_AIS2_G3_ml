Problem solving 2

Add Two Numbers

```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution(object):
    def addTwoNumbers(self, l1, l2):
        dummy = ListNode(0)
        current = dummy
        carry = 0

        while l1 is not None or l2 is not None:
            x = l1.val if l1 is not None else 0
            y = l2.val if l2 is not None else 0
            
            sum_val = carry + x + y  
            carry = sum_val // 10
            current.next = ListNode(sum_val % 10)
            current = current.next

            if l1 is not None: 
                l1 = l1.next
            if l2 is not None: 
                l2 = l2.next

        if carry > 0:
            current.next = ListNode(carry)

        return dummy.next


def create_linked_list(arr):
    """Helper to create ListNode from list"""
    dummy = ListNode(0)
    current = dummy
    for val in arr:
        current.next = ListNode(val)
        current = current.next
    return dummy.next

def linked_list_to_list(node):
    """Helper to convert ListNode to list"""
    result = []
    while node:
        result.append(node.val)
        node = node.next
    return result
```

```python
# Create the linked lists
l1_node = create_linked_list([2, 4, 3])  # 342 in reverse
l2_node = create_linked_list([5, 6, 4])  # 465 in reverse

# Create solution and test
solution = Solution()
result_node = solution.addTwoNumbers(l1_node, l2_node)

# Convert result to list
print("Result:", linked_list_to_list(result_node))  # [7, 0, 8] = 807
```
