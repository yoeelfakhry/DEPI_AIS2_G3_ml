```python
class Student:

    def __init__(self , name="yoeel" , age = 20):
        self.name = name
        self.age  = age

    def get_info(self):
        return f"Welcome {self.name} and your age is {self.age}"
```

```python
# abstract here is a main method that i will use it in different ways and watch the video again
from abc import ABC , abstractmethod

class Animal(ABC):
    def move(self):
        pass

class Bird(Animal):
    def move(self):
        print("Move from Bird")

class cat(Animal):
    def move(self):
        print ("Move from cat")
```

Encapsulation

```python
# __ attribure called private attribute
class Car:
    # private attribute
    __engine_capacity = "2000cc"

    # private method
    def __start_engine (self):
        print ("Engine started")

c1 = Car()
c1.__start_engine()
```

```python
class Calculator:
    def summation(self , n1 , n2):
        return n1 + n2 
    
    def substraction (self , n1 , n2):
        return n1 - n2
    
    def division (self , n1 , n2):
        return n1 / n2
    
    def multiplication (self , n1 , n2):
        return n1 * n2
    
c1 = Calculator()
c1.multiplication(4,2)
```

Multiple inheritance

```python
class A():
    def do_this(self):
        print ("I am in A")
    
class B(A):
    pass 

class C():
    def do_this(self):
        print("I am in c")
class D (B ,C):
    pass 

x = D()
x.do_this()
```

```python
class A():
    def do_this(self):
        print ("I am in A")
    
class B(A):
    pass 

class C():
    def do_this(self):
        print("I am in c")
class D (B , c):
    pass 

```

```python
from prettytable import PrettyTable

products = [
    {"name":"water" , "price":80.0  , "quantity":1500},
    {"name":"soda"  , "price":130.0 , "quantity":1500},
    {"name":"chips" , "price":75.0  , "quantity":1500},
    {"name":"Bread" , "price":45.0  , "quantity":1500},
    {"name":"eggs"  , "price":200.0 , "quantity":1500}
]
table =  PrettyTable()
table.field_names = ["Product Name" , "Price" , "Quantity"]

for p in products :
    table.add_row([p["name"] , f"{p["price"]:.2f}" , p["quantity"]])

print(table)
total_bills = 0 
while True :
    prodcut_name = input("\n Enter product name or ('exit') to finish)").lower().strip()
    if prodcut_name == 'exit' :
        break 
    # the next function stop search when find what it search for 
    product = next( (p for p in products if p["name"].lower() == prodcut_name) , None)

    if product is None :
        print("Product not found. please try again")
        continue
    try :
        quantity_required = float(input(f"How many {product["name"]} do you want ?"))
        if quantity_required <= 0 :
            print("Quantity must be greater than Zero")
            continue

        if quantity_required > product["quantity"]:
            print ("insufficient quantity")
            continue
        
        discount_percentage = 0.05 
        price_before_discount = quantity_required * product["price"]
        discount_amount       = price_before_discount * discount_percentage
        final_price           = price_before_discount - discount_amount

        product["quantity"] -= quantity_required
        total_bills += final_price

        print (f"Success! price for {product['name']}: ${final_price:.2f} (saved :${final_price:.2f} )")  
        another = input ("Somethind else? (y/n)")
        if another != "y":
            break
    
    except ValueError :
        print("Invalid input Please try again")

print(f"\n Total amount to pay : $ {total_bills:.2f}")
print("thank you")

```
