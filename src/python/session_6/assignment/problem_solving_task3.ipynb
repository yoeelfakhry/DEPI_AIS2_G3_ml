Roman to integer

```python
class Solution(object):
    def romanToInt(self, s:str) -> int:
        """
        Calculate the Roman number and return it with integer number
        :type s: str
        :rtype: int
        """
        s = s.upper()
        
        roman_int = {"I": 1 , "V": 5 , "X":10 , "L":50 , "C":100 , "D":500 , "M":1000}
        result = 0
        for idx , letter in enumerate (s):
            if idx+1 < len(s) and roman_int[s[idx]] < roman_int[s[idx+1]]:
                result -=roman_int[s[idx]]
            else : 
                result += roman_int[letter]
        if result < 0 or result > 3999 :
            return "Invalid Roman numeral range"
        return result


```
