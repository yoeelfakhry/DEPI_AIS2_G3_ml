```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4])
y= np.array([2,2.8,3.5,4.5])
```

```python
w = 0 
b = 0
alpha = 0.01
num_ieration  = 20
```

```python
sse_values = []
for i in range(num_ieration):
    y_hat = w * x + b
    
    derivative_w = 2 * np.sum((y_hat - y)*x)
    derivative_b = 2 * np.sum((y_hat -y)*1)
    
    w-= alpha * derivative_w
    b-= alpha * derivative_b

    sse = np.sum((y_hat - y)**2)
    sse_values.append(sse)
    if (i+1) % 20 == 0:
        print(f"iteration: {i+1}, sse:{sse}")
```

```python
print(f"Optimized param : bais = {b} , slop is {w}")
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(range(num_ieration), sse_values , label = 'sse'  )
plt.xlabel("iteration")
plt.ylabel("SSE")
plt.title("SSE Over Iterarion")
plt.show()
plt.subplot(1,2,2)
plt.scatter(x,y, color ='blue' , label = 'Data points')
plt.plot(x , w*x+b , color = 'red' , label = 'regression line')

plt.show()
```
